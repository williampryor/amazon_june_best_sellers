```python
#import dependencies
import pandas as pd
import numpy as np
```

```python
#save path to dataset to the variable 'best_seller_data'
best_seller_data = "Resources/Amazon_Best_Seller_2021_June.csv"

#read in the csv with pandas and display the header of the file
best_seller_df = pd.read_csv(best_seller_data)
best_seller_df.head(10)
```

```python
#clean the dataframe so that 'No of Sellers' Column only shows numerical values instead of string values
best_seller_df['No of Sellers'] = best_seller_df['No of Sellers'].str.replace(r"[a-zA-Z]",'')

#clean the datafram so that the Rank Column only shows numerical values-- remove pound signs
best_seller_df['Rank'] = best_seller_df['Rank'].str.replace(r"[#]", '')

#create a new dataframe without the product link and ASIN column
reduced_best_seller_df = best_seller_df[["Category", "No of Sellers", "Rank", "Rating", "Reviews Count", "Price"]]

#display result
reduced_best_seller_df.head(20)
```
